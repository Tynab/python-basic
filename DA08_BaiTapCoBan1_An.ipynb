# Bài tập Python cơ bản 1 - Quyết định giao dịch cổ phiếu

So sánh giá của từng ngày với ngày liền trước: bán khi giá tăng từ 7%, mua khi giá giảm từ 5%, còn lại giữ nguyên.

```python
from enum import Enum


# Dữ liệu đầu vào theo đúng danh sách giá trong README.
GIA_THEO_NGAY = [10.1, 9.5, 10.7, 10.8, 10.6, 10.0, 10.2, 9.5]
NGUONG_BAN = 0.07
NGUONG_MUA = -0.05


class HanhDong(Enum):
    # Enum giúp giới hạn kết quả trong ba hành động hợp lệ.
    MUA = "Mua"
    BAN = "Bán"
    GIU = "Giữ nguyên"
```

```python
def xac_dinh_hanh_dong(gia_truoc, gia_sau):
    """Trả về hành động giao dịch dựa trên tỷ lệ thay đổi giá."""
    if gia_truoc <= 0:
        raise ValueError("Giá của ngày trước phải lớn hơn 0.")

    ty_le_thay_doi = (gia_sau - gia_truoc) / gia_truoc
    if ty_le_thay_doi >= NGUONG_BAN:
        return HanhDong.BAN
    if ty_le_thay_doi <= NGUONG_MUA:
        return HanhDong.MUA
    return HanhDong.GIU
```

```python
def main(gia_theo_ngay):
    """In quyết định cho từng cặp ngày liên tiếp trong danh sách giá."""
    if len(gia_theo_ngay) < 2:
        print("Cần ít nhất hai mức giá để so sánh.")
        return

    # zip ghép mỗi giá với giá của ngày kế tiếp, tránh truy cập sai chỉ số.
    for ngay, (gia_truoc, gia_sau) in enumerate(
        zip(gia_theo_ngay, gia_theo_ngay[1:]), start=2
    ):
        ty_le = (gia_sau - gia_truoc) / gia_truoc
        hanh_dong = xac_dinh_hanh_dong(gia_truoc, gia_sau)
        print(
            f"Ngày {ngay}: {gia_truoc:.1f} → {gia_sau:.1f} "
            f"({ty_le:+.2%}) - {hanh_dong.value}"
        )
```

```python
# Chạy toàn bộ ví dụ để tạo output có thể kiểm tra lại.
main(GIA_THEO_NGAY)
```

## Tóm tắt

Chương trình duyệt các cặp ngày liên tiếp, tính tỷ lệ thay đổi thay vì chỉ lấy chênh lệch tuyệt đối và trả về đúng một trong ba hành động. Hàm xử lý cũng từ chối giá ngày trước không hợp lệ để tránh phép chia cho 0.
